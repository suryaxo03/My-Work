```python
import pandas as pd
from tabulate import tabulate
```

```python
filepath = "/content/drive/MyDrive/Bharath Intern/Movie Recommendation/Highest Holywood Grossing Movies.csv"
df_movies = pd.read_csv(filepath)
df_movies.head()
```

```python
df = df_movies.drop(['Unnamed: 0', 'Distributor', 'Budget (in $)',
                     'Domestic Opening (in $)', 'Domestic Sales (in $)', 'International Sales (in $)',
                     'World Wide Sales (in $)'], axis = 1)
df.head()
```

```python
df['License'].isna().sum()
```

```python
df['License']=df['License'].fillna('NaN')
```

```python
def recommend_movies_genre(data, genre_filter=None, top_n=None):
    if genre_filter:
        movies = data[data['Genre'].str.contains(genre_filter, case=False)]
    sort_by = input("Sort by date?(Yes/No): ")
    if sort_by.lower() == 'yes':
      movies = movies.sort_values(by='Year', ascending=False)
    if(movies.empty):
      print(f"No movies for {genre_filter} in your genre.")
      return
    print(tabulate(movies[['Title', 'Movie Info', 'Genre', 'Running Time', 'License', 'Release Date', 'Year']]))
```

```python
desired_genre = input("Which genre do you wanna watch?: ")
opt = input("Do you want to sort by top movies?(Yes/No): ")
if opt.lower() == 'yes':
  isbool = recommend_movies_genre(df, genre_filter=desired_genre, top_n=50)
else:
  isbool = recommend_movies_genre(df, genre_filter=desired_genre)
```
