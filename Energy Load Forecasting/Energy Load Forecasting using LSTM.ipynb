```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.layers import LSTM
keras.__version__
```

```python
weather_features = df[['temperature','humidity', 'visibility','windSpeed', 'pressure', 'cloudCover', 'windBearing', 'precipIntensity','precipProbability']]
energy_use = df['use']
weather_features = weather_features['2016-01-02 05:00:00':'2016-12-02 05:00:00']
weather_features['yesterday_use'] = energy_use['2016-01-01 05:00:00':'2016-12-01 05:00:00'].values
energy_use = df['use']['2016-01-01 05:00:00':'2016-12-01 05:00:00']
weather_features.shape, energy_use.shape
```

```python
x_train = weather_features[:6500]
y_train = energy_use[:6500]

x_test = weather_features[6500:]
y_test = energy_use[6500:]

x_train = np.reshape(x_train.values, (x_train.shape[0], x_train.shape[1], 1))
x_test = np.reshape(x_test.values, (x_test.shape[0], x_test.shape[1], 1))

# y_train = np.reshape(y_train.values, (y_train.shape[0], y_train.shape[1], 1))
# y_test = np.reshape(y_test.values, (y_test.shape[0], y_test.shape[1], 1))

x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
model = Sequential()
model.add(LSTM(units=100, return_sequences=True, input_shape=(x_train.shape[1], 1)))
model.add(Dropout(0.2))
model.add(LSTM(units=100))
model.add(Dropout(0.2))
model.add(Dense(units = 20, activation='relu'))
model.compile(optimizer = 'rmsprop', loss = 'mse')
model.summary()
```

```python
model.fit(x_train, y_train, batch_size=32, epochs=100, validation_split=0.05, verbose=1)
```

```python
test_mse = model.evaluate(x_test, y_test, verbose=1)
print ('\nThe mean squared error (MSE) on the test data set is %.3f over %d test samples.' % (test_mse, len(y_test)))
```

```python
predictions = model.predict(x_test)
plt.figure(figsize=(25,10))
plt.plot(y_test.values, color='black', label='Original Usage')
plt.plot(predictions[:,0], color='green', label='Predicted Usage')
plt.title('Energy Usage Prediction')
plt.xlabel('Date')
plt.ylabel('kW')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(25,10))
plt.plot(y_test.values[::24], color='black', label='Original Usage')
plt.plot(predictions[:,0][::24], color='green', label='Predicted Usage')
plt.title('Energy Usage Prediction')
plt.xlabel('Date')
plt.ylabel('kW')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import r2_score

# Assuming you have your actual target values (y_true) and predicted values (y_pred)
r2 = r2_score(y_test, predictions[:, 0])

print(f"R-squared score: {r2*100} %")
```
